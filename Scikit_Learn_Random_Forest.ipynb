Now to convert to numeric datatype

Ordinal encoder

```python
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd

# Sample DataFrame (Ensure df1 is already defined before this step)
# df1 = pd.read_csv('your_file.csv')  # Load your dataset

# Define categorical columns to be encoded
categorical_cols = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']

# Initialize OrdinalEncoder
oe = OrdinalEncoder()

# Fit and transform all categorical columns at once
df1[categorical_cols] = oe.fit_transform(df1[categorical_cols])

# Check the transformed DataFrame
print(df1.head())
```

```python
df1
```

```python
X = df1.iloc[:,0:-1]
y = df.iloc[:,-1]
```

Split data into Training and testing

```python
from sklearn.model_selection import train_test_split

# Ensure X and y are correctly defined
X = df1.drop(columns=['class'])  # Features
y = df1['class']  # Target variable

# Train-Test Split (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)  # Check split sizes
```

```python
from sklearn.ensemble import RandomForestClassifier

clf1 = RandomForestClassifier()
clf1.fit(X_train, y_train)
```

prediction

```python
pred1 = clf1.predict(X_test)
pred1
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred1)
```

```python
from yellowbrick.model_selection import validation_curve

num_est = [100,200,450,500,750,1000]
print(validation_curve(RandomForestClassifier(),X = X_train, y = y_train, param_name='n_estimators', param_range=num_est, scoring ='accuracy', cv = 3))
```

```python
from yellowbrick.model_selection import validation_curve

depth_val = [10,7,  15,20,22]
print(validation_curve(RandomForestClassifier(),X = X_train, y = y_train, param_name='max_depth', param_range=depth_val, scoring ='accuracy', cv = 3))
```

```python
from yellowbrick.model_selection import validation_curve

min_samples = [3,5,4,5,8,12,15]
print(validation_curve(RandomForestClassifier(),X = X_train, y = y_train, param_name='min_samples_split', param_range=min_samples, scoring ='accuracy', cv = 3))
```

```python
clf2 = RandomForestClassifier(n_estimators=1000,min_samples_split = 3, max_depth=15, random_state=0)
clf2.fit(X_train, y_train)

pred2 = clf2.predict(X_test)
accuracy_score(y_test, pred2)
```

Now to Extract features for our use

```python
feature_score= pd.Series(clf2.feature_importances_, index=X_train.columns).sort_values(ascending=False)
```

```python
feature_score
```

```python
sns.barplot(x=feature_score, y=feature_score.index)
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title("Visualizing Important Features")
```

We will remove least performing features for improving accuracy

```python
clf3 = RandomForestClassifier()
Xn = df1.drop(['doors','lug_boot','maint'],axis = 1)
yn = df1['class']

X_train, X_test, y_train, y_test = train_test_split(Xn, yn, test_size=0.3, random_state=0)

clf3.fit(X_train, y_train)

new_pred = clf3.predict(X_test)
accuracy_score(y_test, new_pred)
```
